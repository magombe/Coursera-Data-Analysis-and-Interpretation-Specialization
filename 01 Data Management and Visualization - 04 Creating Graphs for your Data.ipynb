# Data Management and Visualization
### Week 4: Creating Graphs for your Data

In this next stage, I will graph my data in order to visualize the trends.

The first step is to import the data and create a new dataframe with the binned variables.

```python
'''
Code for Peer-graded Assignment: Creating graphs for your data
Course: Data Management and Visualization
Specialization: Data Analysis and Interpretation
'''
 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('gapminder.csv', low_memory=False)
 
bindata = data.copy()
 
data['internetuserate'] = pd.to_numeric(data['internetuserate'], errors='coerce')
bindata['internetuserate'] = pd.cut(data.internetuserate, 10)
 
data['incomeperperson'] = pd.to_numeric(data['incomeperperson'], errors='coerce')
bindata['incomeperperson'] = pd.cut(data.incomeperperson, 10)
 
data['employrate'] = pd.to_numeric(data['employrate'], errors='coerce')
bindata['employrate'] = pd.cut(data.employrate, 10)
 
data['femaleemployrate'] = pd.to_numeric(data['femaleemployrate'], errors='coerce')
bindata['femaleemployrate'] = pd.cut(data.femaleemployrate, 10)
 
data['polityscore'] = pd.to_numeric(data['polityscore'], errors='coerce')
bindata['polityscore'] = data['polityscore']
sub2 = bindata.copy()
```

Plot univariate histograms for the quantitative variables

```python
plt.figure()
sns.distplot(data["internetuserate"].dropna(), kde=False);
plt.xlabel('Internet Use Rate')
plt.title('Percentage of Population with Access to the Internet')
sns.plt.show()
```

The graph is basically unimodal with some noise in it. It is right skewed, displaying the fact that most of the world lacks internet access. There is a small peak around 40%.

```python
plt.figure()
sns.distplot(data["incomeperperson"].dropna(), kde=False);
plt.xlabel('Income Per Person')
plt.title('GDP Per Person')
sns.plt.show()
```

The graph is unimodal and highly right skewed. Incomes are very small for most of the world and the wealthy tail is very long, with a maximum value of more than $100,000.

```python
plt.figure()
sns.distplot(data["employrate"].dropna(), kde=False);
plt.xlabel('Employment Rate')
plt.title('Percentage of total population, age above 15, that has been employed during the given year.')
sns.plt.show()
```

The graph is unimodal and seemingly approaching a normal distribution with a median in the high 50s.

```python
plt.figure()
sns.distplot(data["femaleemployrate"].dropna(), kde=False);
plt.xlabel('Female Employment Rate')
plt.title('Percentage of Female Population with Access to the Internet')
sns.plt.show()
```

```python
plt.figure()
sns.countplot(x="polityscore", data=data)
plt.xlabel('Polity Score')
plt.title("The summary measure of a country's democratic and free nature")
sns.plt.show()
```

Plot basic quantitative to quantitative scatterplots

```python
plt.figure()
scat1 = sns.regplot(x="internetuserate", y="incomeperperson", fit_reg=False, data=data)
plt.xlabel('Internet Use Rate')
plt.ylabel('Income Per Person')
plt.title('Scatterplot for the Association Between Internet Use Rate and Income Per Person')
sns.plt.show()
```

The scatterplot showing the relationship between internet use rate and income per person seems to be exponential, with income per person going up with internet use rate and going up an an accelerating rate.

```python
plt.figure()
scat2 = sns.regplot(x="internetuserate", y="employrate", fit_reg=False, data=data)
plt.xlabel('Internet Use Rate')
plt.ylabel('Employment Rate')
plt.title('Scatterplot for the Association Between Internet Use Rate and Employment Rate')
sns.plt.show()
```

The scatterplot of internet use rate vs. employment rate however, is not so clear. It seems that there may be an upward trend after 60% internet penetration, but at less than 50% internet penetration, there appears to be no relationship.

```python
plt.figure()
scat3 = sns.regplot(x="internetuserate", y="femaleemployrate", fit_reg=False, data=data)
plt.xlabel('Female Employment Rate')
plt.ylabel('Income Per Person')
plt.title('Scatterplot for the Association Between Internet Use Rate and Female Employment Rate')
sns.plt.show()
```

Plot bivariate bar graphs for categorical to quantitative variable analysis

```python
sns.factorplot(x="polityscore", y="internetuserate", data=data, kind="bar", aspect=2, ci=None)
plt.xlabel('Polity Score')
plt.ylabel('Internet Use Rate')
sns.plt.show()
print ()

sns.factorplot(x="polityscore", y="incomeperperson", data=data, kind="bar", aspect=2, ci=None)
plt.xlabel('Polity Score')
plt.ylabel('Income Per Person')
sns.plt.show()
print ()

sns.factorplot(x="polityscore", y="employrate", data=data, kind="bar", aspect=2, ci=None)
plt.xlabel('Polity Score')
plt.ylabel('Employment Rate')
sns.plt.show()
print ()

sns.factorplot(x="polityscore", y="femaleemployrate", data=data, kind="bar", aspect=2, ci=None)
plt.xlabel('Polity Score')
plt.ylabel('Female Employment Rate')
sns.plt.show()
print ()
```
